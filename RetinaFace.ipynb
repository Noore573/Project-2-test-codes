```python
from retinaface import RetinaFace
```

```python
imgpath="D:/Retinafacetest/im.jpg"
resp = RetinaFace.detect_faces(imgpath)
```

```python
resp
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
imgread1re=cv2.imread(imgpath)
len(resp.keys())
```

```python
for key in resp.keys():
    identity=resp[key]
    print(identity)
    facialarea=identity["facial_area"]
    # cv2.rectangle(imgread,(facialarea11111[2],facialarea[3],facialarea[0],facialarea[1]),(255,255,255),1)
    cv2.rectangle(imgread1re, (facialarea[0], facialarea[1]), (facialarea[2], facialarea[3]), (255, 255, 255), 1)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(imgread1re[:,:,::-1])
plt.show()
```
